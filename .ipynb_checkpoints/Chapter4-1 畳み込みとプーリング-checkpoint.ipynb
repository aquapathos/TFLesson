# 畳み込みフィルタの実験

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.examples.tutorials.mnist import input_data
import seaborn as sns
import tensorflow as tf
import pickle
%matplotlib inline
from tensorflow.contrib.deprecated import histogram_summary
```

```python
# テキストファイルから文字列を読み込む
with open('ORENIST.data','r', encoding='utf-8') as a_file:
  txt = a_file.read()
# 読み込んだ文字列のCR(\r)を削除する。
txt = txt.replace('\r', '')
# ファイルをバイナリモードで開く
with open('DANIST.data','wb') as a_file:
  # 文字列をバイト列にして保存する
  a_file.write(txt.encode('utf-8'))
with open('DANIST.data', 'rb') as file:
    images, labels = pickle.load(file, encoding='latin1')
```

```python
fig = plt.figure(figsize=(10,5))
for i in range(40):
    subplot = fig.add_subplot(4,10,i+1)
    subplot.set_xticks([])
    subplot.set_yticks([])
    subplot.set_title('{}'.format(np.argmax(labels[i])))
    subplot.imshow(images[i].reshape(28,28), vmin=0,vmax=1,
                       cmap=plt.cm.binary, interpolation='nearest')
```

```python
def edge_filter():
    vertical = np.array(
        [[2,3,4,3,2],
         [1,2,3,2,1],
         [0,0,0,0,0],
         [-1,-2,-3,-2,-1],
         [-2,-3,-4,-3,-2]]) / 12.0
    horizontal = vertical.T
    matrix = np.zeros([5,5,1,2])
    matrix[:,:,0,0] = vertical
    matrix[:,:,0,1] = horizontal
    
    return tf.constant(matrix, dtype=tf.float32)
```

# 畳み込みネットワークの定義

```python
x = tf.placeholder(tf.float32, [None, 784])
x_image = tf.reshape(x, [-1,28,28,1]) #  28 x 28 x depth 1 、-1 は 枚数が x のサイズに合わせて決まるという意味

W_conv = edge_filter()
h_conv = tf.abs(tf.nn.conv2d(x_image, W_conv, strides = [1,1,1,1], padding='SAME'))

h_conv_cutoff = tf.nn.relu(h_conv-0.5)

h_pool = tf.nn.max_pool(h_conv_cutoff, ksize=[1,4,4,1],      
                       strides=[1,2,2,1], padding='SAME')  # ksize : ブロックサイズ、strides 幾つずつずらすか
```

# セッションの定義

```python
sess = tf.InteractiveSession()
sess.run(tf.global_variables_initializer())
```
