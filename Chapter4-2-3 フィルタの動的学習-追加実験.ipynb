#  フィルタの学習
フィルタを変数化して最適化する

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.examples.tutorials.mnist import input_data
import seaborn as sns
import tensorflow as tf
import pickle
%matplotlib inline
from tensorflow.contrib.deprecated import histogram_summary
np.random.seed(201708)
tf.set_random_seed(201708)
```

# 実験データの読み込み
今回の実験は　中井悦司 著「TensorFlowで学ぶディープラーニング入門～畳み込みニューラルネットワーク徹底解説」で公開されているデータ　https://github.com/enakai00/jupyter_tfbook　を使うが、そのまま pickle で読むとエラーが出たのでまず、変換する。

```python
# テキストファイルから文字列を読み込む
with open('ORENIST.data','r', encoding='utf-8') as a_file:
  txt = a_file.read()
# 読み込んだ文字列のCR(\r)を削除する。
txt = txt.replace('\r', '')
# ファイルをバイナリモードで開く
with open('DANIST.data','wb') as a_file:
  # 文字列をバイト列にして保存する
  a_file.write(txt.encode('utf-8'))
with open('DANIST.data', 'rb') as file:
    images, labels = pickle.load(file, encoding='latin1')
```

# 畳み込みネットワークの定義

```python
x = tf.placeholder(tf.float32, [None, 784])
x_image = tf.reshape(x, [-1,28,28,1]) #  28 x 28 x depth 1 、-1 は 枚数が x のサイズに合わせて決まるという意味

n_filters = 2
W_conv = tf.Variable(tf.truncated_normal([5,5,1,n_filters],stddev=0.008))
h_conv = tf.nn.conv2d(x_image, W_conv, strides = [1,1,1,1], padding='SAME')
b_conv = tf.Variable(tf.zeros([1,n_filters]))
h_conv_cutoff = tf.nn.relu(h_conv - b_conv)

h_pool = tf.nn.max_pool(h_conv_cutoff, ksize=[1,4,4,1],      
                       strides=[1,2,2,1], padding='SAME')  # ksize : ブロックサイズ、strides 幾つずつずらすか
```

# 識別ネットワーク

```python
# n_units = 784 * 2
# input = tf.reshape(h_conv_cutoff, [-1,n_units]) 
n_units = 392
input = tf.reshape(h_pool, [-1,n_units])  # 1次元データに変換 14x14x2 = 392

w = tf.Variable(tf.truncated_normal([n_units,n_filters]))
b = tf.Variable(tf.truncated_normal([n_filters]))
h = tf.nn.relu(tf.matmul(input,w)+b)
w0 = tf.Variable(tf.zeros([n_filters,3]))
b0 = tf.Variable(tf.zeros([3]))
f = tf.matmul(h,w0) + b0
p = tf.nn.softmax(f)
```

```python
ideal = tf.placeholder(tf.float32, [None, 3])
loss = -tf.reduce_sum(ideal* tf.log(p))
train_step = tf.train.AdamOptimizer().minimize(loss)
correct_prediction = tf.equal(tf.argmax(p,1), tf.argmax(ideal,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

# フィルタリング結果の表示間数

```python
def showConv(images,labels,conv_vals,pool_val,filter_vals):
    fig = plt.figure(figsize=(13,5))
    for i in range(2):
        subplot = fig.add_subplot(5,11,11*(i+1)+1)
        subplot.set_xticks([])
        subplot.set_yticks([])
        subplot.imshow(filter_vals[:,:,0,i], vmin=-0.3,vmax=0.3,
                       cmap=plt.cm.bwr, interpolation='nearest')
        
    maxval = np.max(conv_vals) # エッジ強度の最大値
        
    for i in range(10):
        subplot = fig.add_subplot(5,11,i+2)
        subplot.set_xticks([])
        subplot.set_yticks([]);
        subplot.set_title('{}'.format(np.argmax(labels[i])))
        subplot.imshow(images[i].reshape(28,28), vmin=0,vmax=1,
                       cmap=plt.cm.binary, interpolation='nearest')

        subplot = fig.add_subplot(5,11,11+i+2)
        subplot.set_xticks([])
        subplot.set_yticks([]);
        subplot.imshow(conv_vals[i,:,:,0], vmin=0,vmax=maxval,
                       cmap=plt.cm.binary, interpolation='nearest')

    
        subplot = fig.add_subplot(5,11,22+i+2)
        subplot.set_xticks([])
        subplot.set_yticks([]);
        subplot.imshow(conv_vals[i,:,:,1], vmin=0,vmax=maxval,
                       cmap=plt.cm.binary, interpolation='nearest')

        subplot = fig.add_subplot(5,11,33+i+2)
        subplot.set_xticks([])
        subplot.set_yticks([]);
        subplot.imshow(pool_vals[i,:,:,0], vmin=0,vmax=maxval,
                       cmap=plt.cm.binary, interpolation='nearest')
    
        subplot = fig.add_subplot(5,11,44+i+2)
        subplot.set_xticks([])
        subplot.set_yticks([]);
        subplot.imshow(pool_vals[i,:,:,1], vmin=0,vmax=maxval,
                       cmap=plt.cm.binary, interpolation='nearest')  
```

# 学習セッション

```python
sess = tf.InteractiveSession()
sess.run(tf.global_variables_initializer())
```

```python
filter_vals, conv_vals, pool_vals = sess.run([W_conv, h_conv_cutoff,h_pool],
                            feed_dict={x:images[30:40]})

showConv(images[10:20],labels[10:20],conv_vals,pool_vals,filter_vals)
```

```python
times = 0
for _ in range(200):
    times += 1
    sess.run(train_step, feed_dict = {x:images, ideal:labels})
    if times  % 20 == 0:
        loss_val, acc_val = sess.run(
            [loss, accuracy], feed_dict = {x:images, ideal:labels})
        print("Step: {0:d}, Loss: {1:f}, Accuracy: {2:f}".format(times, loss_val, acc_val))
```

```python
filter_vals, conv_vals, pool_vals = sess.run([W_conv, h_conv_cutoff,h_pool],
                            feed_dict={x:images[10:20]})

showConv(images[10:20],labels[10:20],conv_vals,pool_vals,filter_vals)
```

```python
def showresult(start,sess):
    fig = plt.figure(figsize=(13,5))
    img = images[start:start+10]
    lbl = labels[start:start+10]
    p_val = sess.run(p, feed_dict={x:img})
    for i in range(10):
        subplot = fig.add_subplot(1,10,i+1)
        subplot.set_xticks([])
        subplot.set_yticks([])
        subplot.set_title("{}/{}".format(np.argmax(lbl[i]),np.argmax(p_val[i])))
        subplot.imshow(img[i].reshape(28,28), vmin=0,vmax=1,
                       cmap=plt.cm.binary, interpolation='nearest')

def showH(sess):
    hidden =  sess.run(
            h, feed_dict = {x:images, ideal:labels})

    u = [[],[],[]]
    v = [[],[],[]]

    for i in range(90):
        m = np.argmax(labels[i])
        h1 = hidden[i]
        u[m].append(h1[0])
        v[m].append(h1[1])
    
    fig = plt.figure(figsize=(5,5))
    subplot = fig.add_subplot(1,1,1)
    subplot.scatter(u[0],v[0],s=200, marker='|')
    subplot.scatter(u[1],v[1],s=200, marker='_')
    subplot.scatter(u[2],v[2],s=200, marker='.')

showresult(10,sess)
showresult(50,sess)
```

```python
sess.close()
x = tf.placeholder(tf.float32, [None, 784])
x_image = tf.reshape(x, [-1,28,28,1]) #  28 x 28 x depth 1 、-1 は 枚数が x のサイズに合わせて決まるという意味

n_filters = 2
W_conv = tf.Variable(tf.truncated_normal([5,5,1,n_filters],stddev=0.002))
h_conv = tf.nn.conv2d(x_image, W_conv, strides = [1,1,1,1], padding='SAME')
b_conv = tf.Variable(tf.zeros([1,n_filters]))
h_conv_cutoff = tf.nn.relu(h_conv - b_conv)

h_pool = tf.nn.max_pool(h_conv_cutoff, ksize=[1,4,4,1],      
                       strides=[1,2,2,1], padding='SAME')  # ksize : ブロックサイズ、strides 幾つずつずらすか

n_units = 392
input = tf.reshape(h_pool, [-1,n_units])  # 1次元データに変換 14x14x2 = 392

w = tf.Variable(tf.truncated_normal([n_units,n_filters]))
b = tf.Variable(tf.truncated_normal([n_filters]))
h = tf.nn.relu(tf.matmul(input,w)+b)
w0 = tf.Variable(tf.zeros([n_filters,3]))
b0 = tf.Variable(tf.zeros([3]))
f = tf.matmul(h,w0) + b0
p = tf.nn.softmax(f)

ideal = tf.placeholder(tf.float32, [None, 3])
loss = -tf.reduce_sum(ideal* tf.log(p))
train_step = tf.train.AdamOptimizer().minimize(loss)
correct_prediction = tf.equal(tf.argmax(p,1), tf.argmax(ideal,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

sess = tf.Session()
sess.run(tf.global_variables_initializer())
sess.run(tf.local_variables_initializer())
times = 0
for _ in range(200):
    times += 1
    sess.run(train_step, feed_dict = {x:images, ideal:labels})
    if times  % 20 == 0:
        loss_val, acc_val = sess.run(
            [loss, accuracy], feed_dict = {x:images, ideal:labels})
        print("Step: {0:d}, Loss: {1:f}, Accuracy: {2:f}".format(times, loss_val, acc_val))
```
