```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.examples.tutorials.mnist import input_data
# import seaborn as sns
%matplotlib inline
```

```python
def testplot():
    flip=1
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)

    plt.show()
    
testplot()
```
